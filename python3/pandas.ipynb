## 3, Series 对象

pandas 使用 Series 对象来表现行和列，当从 dataframe 中提取一行时，pandas 返回一个 Series 对象，其中不仅包含了当前行的值，同时包含列的元信息：

```
NDB_No                  1001
Shrt_Desc   BUTTER WITH SALT
Water_(g)              15.87
...                      ...
```

````python
# Series object representing the row at index 0.
row = food_info.loc[0]
print(type(row))
print(row)
````

### 3.1, pandas 的切片操作同时包含首尾数据

````python
# DataFrame containing the rows at index 3, 4, 5, and 6 returned.
food_info.loc[3:6]
````

````python
# DataFrame containing the rows at index 2, 5, and 10 returned. Either of the following work.
# Method 1
two_five_ten = [2,5,10] 
food_info.loc[two_five_ten]

# Method 2
food_info.loc[[2,5,10]]
````

## 4, dtype

当从一个文件加载 dataframe 时，pandas 会推测每列的类型。可以使用 DataFrame.dtypes 来获取一个 Series 对象，它反映了列名和对应的类型。


pandas 借用了 NumPy 的类型定义，包含以下几种 dtypes：

- object - for representing string values.
- int - for representing integer values.
- float - for representing float values.
- datetime - for representing time values.
- bool - for representing Boolean values.

````python
print(food_info.dtypes)
````
